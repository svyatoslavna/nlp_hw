# Токенизация в NLP

**Токенизация** – процесс разбиения текста на минимальные единицы (токены): слова, подслова или символы


## 1. Простые методы токенизации


### 1.2. Регулярные выражения


```python
import re

text = "Hello, world! This is a test."
tokens = re.findall(r"\w+", text)
tokens
```

- Не учитывают пунктуацию

## 5. Пример предобработки текста

Возьмём новостной текст и применим полный пайплайн предобработки: очистка → токенизация → лемматизация → удаление стоп-слов


```python
import spacy
from nltk.corpus import stopwords
import nltk
import re

nltk.download('stopwords')

# Загрузка модели spaCy для английского
nlp = spacy.load("en_core_web_sm")
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    # 1. Приведение к нижнему регистру
    text = text.lower()
    # 2. Удаление пунктуации и цифр
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # 3. Токенизация и лемматизация через spaCy
    doc = nlp(text)
    # 4. Удаление стоп-слов и коротких токенов
    tokens = [token.lemma_ for token in doc if token.text not in stop_words and len(token.text) > 2]
    return tokens

# Пример
sample_text = "The quick brown fox jumps over the lazy dog. This is a simple sentence for preprocessing!"
processed = preprocess_text(sample_text)
print("Оригинал:", sample_text)
print("После предобработки:", processed)
```
